```python
import pandas as pd
import sqlite3

con = sqlite3.connect("data/nba.sqlite")
games = pd.read_sql_query("SELECT * FROM game LIMIT 100", con)
active_players = pd.read_sql_query("SELECT full_name FROM player WHERE is_active=TRUE", con)

print("Sample Games Data:")
print(games.head())

# Load all active players into a DataFrame
active_players = pd.read_sql_query("SELECT full_name FROM player WHERE is_active=TRUE", con)
print("Active Players:")
print(active_players.head())

# Step 1: Find the most recent season
recent_season = pd.read_sql_query("SELECT MAX(season) AS recent_season FROM game", con)
print(f"Most Recent Season: {recent_season.iloc[0, 0]}")

# Step 2: Get all teams for the most recent season
teams = pd.read_sql_query(f"""
    SELECT DISTINCT t.team_name 
    FROM team t
    JOIN game g ON g.home_team_id = t.team_id OR g.away_team_id = t.team_id
    WHERE g.season = {recent_season.iloc[0, 0]}
    """, con)
print("Teams in the Most Recent Season:")
print(teams)

# Select a team (you can pick any team from the list printed above)
team_name = 'Chosen_Team_Name'  # Replace this with the actual team name

# Step 3: Get Home Game Scores for the Chosen Team
home_games = pd.read_sql_query(f"""
    SELECT g.game_id, g.home_team_id, g.home_team_score 
    FROM game g
    JOIN team t ON g.home_team_id = t.team_id
    WHERE t.team_name = '{team_name}' AND g.season = {recent_season.iloc[0, 0]}
    """, con)

# Step 4: Get Away Game Scores for the Chosen Team
away_games = pd.read_sql_query(f"""
    SELECT g.game_id, g.away_team_id, g.away_team_score
    FROM game g
    JOIN team t ON g.away_team_id = t.team_id
    WHERE t.team_name = '{team_name}' AND g.season = {recent_season.iloc[0, 0]}
    """, con)

# Step 5: Calculate the Average Home and Away Scores
home_avg_score = home_games['home_team_score'].mean()
away_avg_score = away_games['away_team_score'].mean()

print(f"Average Home Score for {team_name}: {home_avg_score}")
print(f"Average Away Score for {team_name}: {away_avg_score}")
```
