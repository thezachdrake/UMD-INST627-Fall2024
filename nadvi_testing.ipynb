1. How does the team’s win rate change when they lead at halftime compared to when they trail at halftime, and how does this vary based on the halftime score margin?

```python
query_1 = """
WITH halftime_scores AS (
    SELECT 
        game_id,
        CASE 
            WHEN pts_home > pts_away THEN 'Leading'
            WHEN pts_home < pts_away THEN 'Trailing'
            ELSE 'Tied'
        END AS halftime_status,
        ABS(pts_home - pts_away) AS halftime_margin,
        CASE 
            WHEN (pts_home > pts_away AND wl_home = 'W') OR 
                 (pts_home < pts_away AND wl_away = 'W') THEN 1
            ELSE 0
        END AS win
    FROM game
)
SELECT 
    halftime_status,
    CASE 
        WHEN halftime_margin BETWEEN 0 AND 5 THEN '0-5'
        WHEN halftime_margin BETWEEN 6 AND 10 THEN '6-10'
        ELSE '10+'
    END AS margin_range,
    COUNT(*) AS games_played,
    AVG(win) AS win_rate
FROM halftime_scores
GROUP BY halftime_status, margin_range;
"""
```

```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("data/nba.sqlite")

# Execute the query
result = pd.read_sql_query(query_1, con)

# Close the connection
con.close()

# Display the result
print("Results for Question 1:")
print(result)
```

Understanding these relationships can provide valuable insights into the importance of halftime performance in basketball.

3. What impact does the timing and frequency of three-point attempts per quarter have on the team’s win probability in different game situations (e.g., trailing, tied, or leading)?

```python
query_3 = """
WITH three_point_attempts AS (
    SELECT 
        game_id,
        period,
        scoremargin,
        CASE 
            WHEN CAST(scoremargin AS INTEGER) > 0 THEN 'Leading'
            WHEN CAST(scoremargin AS INTEGER) < 0 THEN 'Trailing'
            ELSE 'Tied'
        END AS game_state
    FROM play_by_play
    WHERE 
        homedescription LIKE '%3PT%' OR 
        visitordescription LIKE '%3PT%' OR 
        neutraldescription LIKE '%3PT%'
),
three_point_summary AS (
    SELECT
        period,
        game_state,
        COUNT(*) AS attempts
    FROM three_point_attempts
    GROUP BY period, game_state
)
SELECT
    period,
    game_state,
    attempts AS total_attempts
FROM three_point_summary;
"""
```

```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("data/nba.sqlite")

# Execute the corrected query and load the result into a DataFrame
result = pd.read_sql_query(query_3, con)

# Close the database connection
con.close()

# Display the results
print("Results for Question 3:")
print(result)
```

4. How does the home court advantage affect team performance in terms of average points scored, win rate, and scoring margin across all games?

```python

# Define the query
query_4 = """
SELECT 
    'Home' AS location,
    AVG(pts_home) AS avg_points,
    AVG(CASE WHEN wl_home = 'W' THEN 1 ELSE 0 END) AS win_rate,
    AVG(pts_home - pts_away) AS avg_scoring_margin
FROM game
UNION ALL
SELECT 
    'Away' AS location,
    AVG(pts_away) AS avg_points,
    AVG(CASE WHEN wl_home = 'L' THEN 1 ELSE 0 END) AS win_rate,
    AVG(pts_away - pts_home) AS avg_scoring_margin
FROM game;
"""
```

```python
import pandas as pd
import sqlite3

# Connect to the SQLite database
con = sqlite3.connect("data/nba.sqlite")

# Execute the corrected query and load the result into a DataFrame
result = pd.read_sql_query(query_4, con)

# Close the database connection
con.close()

# Display the results
print("Results for Question 4:")
print(result)
```

Intro: Home court advantage is often cited as a critical factor in basketball. This analysis explores its impact on team performance by comparing average points scored, win rates, and scoring margins for home and away games.

Rising Action: We categorized games based on location (Home or Away) and calculated three key metrics:

Average Points Scored: How many points teams score on average.
Win Rate: The percentage of games won.
Scoring Margin: The difference between points scored and allowed.
Climax:
The results reveal a strong home court advantage:

Home Teams: Average 104.62 points per game, a win rate of 61.87%, and a scoring margin of +3.63 points.
Away Teams: Average 100.99 points per game, a win rate of 38.12%, and a scoring margin of -3.63 points.

Falling Action: Home teams consistently outperform away teams across all metrics. The higher scoring margin and win rate at home underscore the psychological and logistical benefits of playing on familiar turf.

Conclusion: The home court advantage significantly enhances team performance, emphasizing its importance in competitive basketball. Teams should leverage this edge by optimizing strategies for home games.

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare data for visualization
categories = ['Average Points', 'Win Rate', 'Scoring Margin']
home_values = [104.62, 0.618725, 3.627569]
away_values = [100.99, 0.381244, -3.627569]
x = np.arange(len(categories))

# Create the grouped bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
plt.bar(x - bar_width/2, home_values, bar_width, label='Home', color='blue')
plt.bar(x + bar_width/2, away_values, bar_width, label='Away', color='orange')

# Add labels, title, and legend
plt.xlabel("Metrics", fontsize=14)
plt.ylabel("Values", fontsize=14)
plt.title("Home vs Away Performance Metrics", fontsize=16)
plt.xticks(x, categories, fontsize=12)
plt.legend(title="Location", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars with values
for i in range(len(home_values)):
    plt.text(x[i] - bar_width/2, home_values[i], f"{home_values[i]:.2f}", ha='center', va='bottom', fontsize=10)
    plt.text(x[i] + bar_width/2, away_values[i], f"{away_values[i]:.2f}", ha='center', va='bottom', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
