```python
import pandas as pd
import sqlite3

con = sqlite3.connect("data/nba.sqlite")
games = pd.read_sql_query("SELECT * FROM game LIMIT 100", con)
active_players = pd.read_sql_query("SELECT full_name FROM player WHERE is_active=TRUE", con)
```

```python
new_orleans_games = pd.read_sql_query("SELECT team_abbreviation_home, pts_home, team_abbreviation_away, pts_away, season_id, game_date FROM game WHERE team_abbreviation_home='NOP' OR team_abbreviation_away='NOP' ORDER BY game_date DESC LIMIT 10", con)
new_orleans_games
```

```python
nop_home_avg = pd.read_sql_query("SELECT team_name_home, avg(pts_home) AS avg_home_pts FROM game WHERE team_abbreviation_home='NOP' AND season_id=22022", con)
nop_home_avg
```

```python
nop_away_avg = pd.read_sql_query("SELECT team_name_away, avg(pts_away) AS avg_away_pts FROM game WHERE team_abbreviation_away='NOP' AND season_id=22022", con)
nop_away_avg
```

```python
from statistics import mean 
print("Overall Average Score: " + str(mean([nop_home_avg["avg_home_pts"][0], nop_away_avg["avg_away_pts"][0]])))
```
