```python
import pandas as pd
import sqlite3

con = sqlite3.connect("data/nba.sqlite")
games = pd.read_sql_query("SELECT * FROM game order by game_date DESC LIMIT 100", con)

#adding the order by clause lets me see what the latest game was and pick Miami Heat
active_players = pd.read_sql_query("SELECT full_name FROM player WHERE is_active=TRUE", con)

games
```

```python
#Team chosen: Miami Heat
#Season: 42022
miamiheat_homegames = pd.read_sql_query("SELECT * FROM game WHERE team_name_home = 'Miami Heat' and season_id = 42022 ", con)
miamiheat_awaygames = pd.read_sql_query("SELECT * FROM game WHERE team_name_away ='Miami Heat'and season_id = 42022", con)
#Compute average points of home and away games
miamiheat_homegames
homeptsavg = miamiheat_homegames.loc[:,'pts_home'].mean()
awayptsavg = miamiheat_awaygames.loc[:,'pts_away'].mean()

print("Miami Heat Average Home PTS:", homeptsavg)
print("Miami Heat Average AWAY PTS:", awayptsavg)
```
