1. How does the team’s win rate change when they lead at halftime compared to when they trail at halftime, and how does this vary based on the halftime score margin?

```python
query_1 = """
WITH halftime_scores AS (
    SELECT 
        game_id,
        CASE 
            WHEN pts_home > pts_away THEN 'Leading'
            WHEN pts_home < pts_away THEN 'Trailing'
            ELSE 'Tied'
        END AS halftime_status,
        ABS(pts_home - pts_away) AS halftime_margin,
        CASE 
            WHEN (pts_home > pts_away AND wl_home = 'W') OR 
                 (pts_home < pts_away AND wl_away = 'W') THEN 1
            ELSE 0
        END AS win
    FROM game
)
SELECT 
    halftime_status,
    CASE 
        WHEN halftime_margin BETWEEN 0 AND 5 THEN '0-5'
        WHEN halftime_margin BETWEEN 6 AND 10 THEN '6-10'
        ELSE '10+'
    END AS margin_range,
    COUNT(*) AS games_played,
    AVG(win) AS win_rate
FROM halftime_scores
GROUP BY halftime_status, margin_range;
"""
```

2. How does the timing and type of player substitutions during the third and fourth quarters impact the team’s scoring differential in close games (score margin of 5 points or less)?

```python
query_2 = """
SELECT 
    game_id,
    CASE 
        WHEN ABS(pts_home - pts_away) <= 5 THEN 'Close Game'
        ELSE 'Not Close'
    END AS game_status,
    AVG(pts_home - pts_away) AS avg_score_differential
FROM game
GROUP BY game_status;
"""
```

3. What impact does the timing and frequency of three-point attempts per quarter have on the team’s win probability in different game situations (e.g., trailing, tied, or leading)?

```python
query_3 = """
WITH three_point_attempts AS (
    SELECT 
        game_id,
        period,
        scoremargin,
        CASE 
            WHEN CAST(scoremargin AS INTEGER) > 0 THEN 'Leading'
            WHEN CAST(scoremargin AS INTEGER) < 0 THEN 'Trailing'
            ELSE 'Tied'
        END AS game_state
    FROM play_by_play
    WHERE 
        homedescription LIKE '%3PT%' OR 
        visitordescription LIKE '%3PT%' OR 
        neutraldescription LIKE '%3PT%'
),
three_point_summary AS (
    SELECT
        period,
        game_state,
        COUNT(*) AS attempts
    FROM three_point_attempts
    GROUP BY period, game_state
)
SELECT
    period,
    game_state,
    attempts AS total_attempts
FROM three_point_summary;
"""
```

4. How does the home court advantage affect team performance in terms of average points scored, win rate, and scoring margin across all games?

```python

# Define the query
query_4 = """
SELECT 
    'Home' AS location,
    AVG(pts_home) AS avg_points,
    AVG(CASE WHEN wl_home = 'W' THEN 1 ELSE 0 END) AS win_rate,
    AVG(pts_home - pts_away) AS avg_scoring_margin
FROM game
UNION ALL
SELECT 
    'Away' AS location,
    AVG(pts_away) AS avg_points,
    AVG(CASE WHEN wl_home = 'L' THEN 1 ELSE 0 END) AS win_rate,
    AVG(pts_away - pts_home) AS avg_scoring_margin
FROM game;
"""
```
