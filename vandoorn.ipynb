### 10/7 Join Crimes and Arrests

```python
arrests = pd.read_csv("data/arrests.csv")
crimes = pd.read_csv("data/crimes.csv")
```

```python
arrests.head()
```

```python
crimes.head()
```

```python
print(f'crime_cols:{crimes.columns.tolist()}, arrest_cols:{arrests.columns.tolist()}')
```

```python
# Merge arrest and crime data on relevant columns
crime_arrest_enriched = arrests.merge(crimes, 
                                          on=['PD_CD', 'KY_CD', 'JURISDICTION_CODE', 'Latitude', 'Longitude'], 
                                          suffixes=('', '_crime'))

crime_arrest_enriched = crime_arrest_enriched[['ARREST_KEY', 'ARREST_DATE', 'PD_CD', 'KY_CD', 'OFNS_DESC', 
                                               'LAW_CAT_CD', 'ARREST_BORO', 'ARREST_PRECINCT', 'Latitude', 
                                               'Longitude', 'CMPLNT_NUM', 'BORO_NM', 'CMPLNT_FR_DT', 
                                               'CMPLNT_FR_TM', 'LAW_CAT_CD_crime', 'OFNS_DESC_crime', 
                                               'PREM_TYP_DESC', 'SUSP_RACE', 'SUSP_SEX', 'VIC_AGE_GROUP', 
                                               'VIC_RACE', 'VIC_SEX']]
```
