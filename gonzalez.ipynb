```python
import pandas as pd
import sqlite3

#Perry Gonzalez, Issue 4
#Find a team that played in the most recent season
con = sqlite3.connect("data/nba.sqlite")
games = pd.read_sql_query("SELECT * FROM game WHERE game_date > '2023-01-01 00:00:00'", con)
games
```

```python
#Compute and display their average home and away points
games = pd.read_sql_query("SELECT 'Miami Heat' AS 'Team Name', AVG(pts_home) AS 'Average Home Points' FROM game WHERE team_name_home = 'Miami Heat'", con)
games2 = pd.read_sql_query("SELECT 'Miami Heat' AS 'Team Name', AVG(pts_away) AS 'Average Away Points' FROM game WHERE team_name_away = 'Miami Heat'", con)
games,games2
```
