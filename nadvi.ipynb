```python
import pandas as pd
import sqlite3
from statistics import mean

con = sqlite3.connect("data/nba.sqlite")

# Retrieving data
games = pd.read_sql_query("SELECT * FROM game LIMIT 100", con)
active_players = pd.read_sql_query("SELECT full_name FROM player WHERE is_active=TRUE", con)
```

```python
# Los Angeles Lakers games filter
lakers_games = pd.read_sql_query("SELECT team_abbreviation_home, pts_home, team_abbreviation_away, pts_away, season_id, game_date FROM game WHERE team_abbreviation_home='LAL' OR team_abbreviation_away='LAL' ORDER BY game_date DESC LIMIT 10", con)
lakers_games
```

```python
# Calculating average home points for Lakers
lal_home_avg = pd.read_sql_query("SELECT team_name_home, ROUND(AVG(pts_home), 2) AS avg_home_pts FROM game WHERE team_name_home='Los Angeles Lakers' AND season_id=22022", con)
lal_home_avg
```

```python
# Calculating average away points for Lakers
lal_away_avg = pd.read_sql_query("SELECT team_name_away, ROUND(AVG(pts_away), 2) AS avg_away_pts FROM game WHERE team_name_away='Los Angeles Lakers' AND season_id=22022", con)
lal_away_avg

# Calculate overall average score
print("Overall Average Score: " + str(round(mean([lal_home_avg["avg_home_pts"][0], lal_away_avg["avg_away_pts"][0]]), 2)))
```
