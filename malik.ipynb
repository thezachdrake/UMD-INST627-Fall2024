```python
#Issue 1 - filtering the data for grand larceny crimes and sorting by location type
import pandas as pd
df = pd.read_csv("data/data.csv")
grand_larceny_crimes = df[df['OFNS_DESC'] == 'GRAND LARCENY']
grand_larceny_crimes = grand_larceny_crimes.sort_values("PREM_TYP_DESC")
grand_larceny_crimes
```

```python
#issue2 - merge the crime data onto the arrest record by age group and gender
import pandas as pd
arrests = pd.read_csv('data/arrests.csv')
crimes = pd.read_csv('data/crimes.csv')
merge_crimes_on_arrest = pd.merge(arrests, crimes, left_on=["KY_CD", "AGE_GROUP","PERP_SEX"], right_on=["KY_CD", "SUSP_AGE_GROUP","SUSP_SEX"], how = "left")
pd.set_option("display.max_columns", None) #to view all columns
merge_crimes_on_arrest
```
