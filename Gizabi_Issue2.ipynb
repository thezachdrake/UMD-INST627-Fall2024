```python
import pandas as pd
crimes = pd.read_csv("data/crimes.csv")
arrests = pd.read_csv("data/arrests.csv")

crimes.info()
```

```python
arrests.info()
```

```python
crimes_and_arrests = arrests.merge(crimes, how = 'left', left_on = ['Longitude', 'Latitude'], right_on = ['Longitude', 'Latitude'], suffixes = ('', '_c'))
crimes_and_arrests.info()
```

```python
enhanced_crimes_and_arrests = crimes_and_arrests[[ 'SUSP_AGE_GROUP', 'SUSP_RACE', 'SUSP_SEX', 'VIC_AGE_GROUP', 'VIC_RACE', 'VIC_SEX']]

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
enhanced_crimes_and_arrests.head()
```
